```python
##create movie recommendation
```

```python
df.head()
```

```python
moviemat=df.pivot_table(index="used_id",columns="title",values="rating")
```

```python
moviemat.head()
```

```python
rating.sort_values('num of ratings',ascending=False).head()
```

```python
StarWars=moviemat['Star Wars (1977)']
```

```python
StarWars.head()
```

```python
similar_to_starwars=moviemat.corrwith(StarWars)
```

```python
corr_starwars=pd.DataFrame(similar_to_starwars,columns=['Correlation'])
corr_starwars.dropna(inplace=True)
```

```python
corr_starwars.head()
```

```python
corr_starwars.sort_values('Correlation', ascending=False).head(10)
```

```python
rating
```

```python
corr_starwars.join(rating['num of ratings'])
```

```python
corr_starwars
```

```python
corr_starwars.head()
```

```python
corr_starwars[corr_starwars['num of ratings']>100].sort_values('Correlation', ascending=False)
```

```python
def predictions(movie_name):
    movie_user_rating=moviemat[movie_name]
    similar_to_movie=moviemat.corrwith(movie_user_rating)
    corr_movie=pd.DataFrame(similar_to_movie,columns=['Correlation'])
    corr_movie.dropna(inplace=True)
    corr_movie=corr_movie.join(rating['num of ratings'])
    pre=corr_movie[corr_movie['num of ratings']>100].sort_values('Correlation', ascending=False)

    return pre


    
```

```python
pre=predictions("Titanic (1997)")
```
